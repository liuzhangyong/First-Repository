### 客户入会时间分布分析

## 相关性分析

提取入会时间、会员卡级别、客户年龄、飞行次数、总飞行公里数、最近一次乘机至观测窗口结束时长、积分兑换次数、总累计积分属性，通过相关系数矩阵与热力图分析各属性之间的相关性。

```python
# 提取相关属性
data_corr = airline_data[['FFP_TIER', 'FLIGHT_COUNT', 'LAST_TO_END', 'SEG_KM_SUM', 'EXCHANGE_COUNT', 'Points_Sum']]
age_fill = airline_data['AGE'].fillna(0)
data_corr['AGE'] = age_fill.astype('int64')
data_corr['ffp_year'] = ffp_year
# 计算皮尔森相关性矩阵
corr_matrix = data_corr.corr(method='pearson')
print('相关性矩阵：\n', corr_matrix)
```

```python
# 绘制热力图
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, vmax=1, square=True, cmap='Blues')
```

# 数据预处理

## 数据清洗

```python
# 去除票价为空的记录
# exp1 = airline_data['SUM_YR_1'].notnull()
# exp2 = airline_data['SUM_YR_2'].notnull()
# exp = exp1 & exp2
# # exp
# airline_notnull = airline_data.loc[exp, :]
airline_notnull = airline_data.dropna(subset=['SUM_YR_1', 'SUM_YR_2'],axis=0) # 删除包含缺失值的行（样本）
print('删除缺失记录后数据的形状：', airline_notnull.shape)
```

```python
# 只保留票价非0，或者平均折扣率不为0且总飞行千米数大于0的记录
index1 = airline_notnull['SUM_YR_1'] != 0
index2 = airline_notnull['SUM_YR_2'] != 0
index3 = (airline_notnull['SEG_KM_SUM'] > 0) & (airline_notnull['avg_discount'] != 0)
airline = airline_notnull[(index1 | index2) & index3]
print('删除异常记录后数据的形状为：', airline.shape)
```

## 特征选取与构造

在这里我将客户关系长度L、消费时间间隔R、消费频率F、飞行里程M和平均折扣系数C这5个特征作为客户价值特征，记为LRFMC模型。

```python
# 选取需求特征
airline_selection = airline[['FFP_DATE', 'LOAD_TIME', 'FLIGHT_COUNT', 'LAST_TO_END', 
                            'avg_discount', 'SEG_KM_SUM']]
# airline_selection
# 构建L特征
L = pd.to_datetime(airline_selection['LOAD_TIME']) - pd.to_datetime(airline_selection['FFP_DATE'])
L = L.astype('str').str.split().str[0]
L = L.astype('int') / 30
# 合并特征
airline_features = pd.concat([L, airline_selection.iloc[:, 2:]], axis=1)
print('构建的LRFMC特征前5行为：\n', airline_features.head())
```

## 数据标准化

目前数据标准化的方法通常有三种，分别是标准差标准化（使用最广泛）、离差标准化、小数定标标准化。
标准差标准化：经过该方法处理的数据均值为0，标准差为1。
离差标准化：对原始数据的一种线性变换，结果是将原始数据映射到[0, 1]区间。
小数定标标准化：通过移动数据的小数位数，将数据映射到区间[-1,1]，移动的小数位数取决于数据绝对值的最大值。

```python
# 这里采用标准差标准化方法
from sklearn.preprocessing import StandardScaler
data = StandardScaler().fit_transform(airline_features)
print('标准化后LRFMC模型的5个特征为：\n', data[:5, :])
```

# 模型构建

```python
# 采用K-Means聚类算法对客户进行分群，尝试5个聚类中心（可以用网格搜索确定最佳聚类中心数）。
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=5, random_state=123).fit(data)
kmean.cluster_centers_  # 查看聚类中心(分类簇的均值向量)
```

```python
kmean.labels_  # 查看样本的类别标签
```

```python
# 统计不同类别样本的数目
r1 = pd.Series(kmean.labels_).value_counts()
print('最终每个类别的数目为：\n', r1)
```
